## Buscar datos del precio de las acciones

```python
# Instalamos e importamos librerías yfinance y matplotlib

# Yahoo finance
import yfinance

# Gráficas
import matplotlib.pyplot as plt
```

```python
# Hacemos uso del dataset de yahoo finance

ticker = input('Ingresa el código de la acción: ')
data = yfinance.Ticker(ticker).history('6mo')
close = data.Close
close
```

```python
# Graficamos el cierre de las acciones de los últimos 6 meses


close.plot()
plt.title('Cierre de acciones APLE (6meses)')
plt.xlabel('Cierre')
plt.ylabel('USD')
plt.savefig("Cierre_acciones_AAPL.jpg", dpi=300, bbox_inches='tight') # 'bbox_inches=tight' ajusta los bordes [5, 15]
plt.show()
```

## Realizar análisis solicitado por el gestor

### Cotización máxima

```python
# buscamos el cierre máximo

close_max = round(close.max(), 2)
print(f'Cierre máximo del periodo: {close_max}')
```

### Cotización mínima

```python
# buscamos el cierre mínimo

close_min = round(close.min(), 2)
print(f'Cierre mínimo del periodo: {close_min}')
```

### Cotización promedio

```python
# buscamos el cierre promedio

close_mean = round(close.mean(), 2)
print(f'Cierre promedio del periodo: {close_mean}')
```

### Generación automática

```python
# Instrucciones para generar el correo de manera automática

# Definimos función copiar/pegar/sleep
def copiar_y_pegar(texto):
    pyperclip.copy(texto)
    pyautogui.hotkey('ctrl', 'v')
    time.sleep(0.5)

# Definimos función para enviar correo automático
def enviar_correo_auto()
    webbrowser.open('https://mail.google.com/mail/u/1/#inbox') # Abrimos navegador
    time.sleep(15) # espera 8 seg
    #pyautogui.click(89,205) # click en botón redactar
    pyautogui.hotkey('d') # abre redactar correo nuevo
    pyautogui.PAUSE = 30
    pyautogui.click(270, 140) # activa la ventana antes de iniciar
    copiar_y_pegar(destinatario) # copiamos destinatario
    pyautogui.hotkey('tab') # se aplica tabulación para pasar a la sig casilla de texto
    copiar_y_pegar(asunto) # copiamos asunto
    pyautogui.hotkey('tab')
    copiar_y_pegar(mensaje) # copiamos mensaje
    pyautogui.click(257, 690) # click en botón adjuntar
    pyautogui.PAUSE = 2
    copiar_y_pegar(ruta_adjunto) # copiamos ruta de archivo
    pyautogui.press('enter') # para aceptar
    pyautogui.PAUSE = 2
    pyautogui.hotkey('ctrl', 'enter') # enviar correo
    pyautogui.PAUSE = 2
    pyautogui.hotkey('ctrl', 'w') # cerramos navegador
    print("Correo enviado exitosamente.") 
    # para una acción más avanzada,aquí se puede aplicar una validación y guardar logs de errores

```
